```python
from pathlib import Path
import json
import pandas as pd
```

```python
data_dir = Path("/home/user/KODC/data")
```

```python
file_list = sorted(data_dir.glob("*.json"))
print(file_list[0:5])
```

```python
for file in file_list:
    fname = file.stem
    with open(file, "r", encoding="utf-8") as file_obj:
        json_data = json.load(file_obj)["body"]["item"]
        df = pd.DataFrame(json_data)
        df.to_csv(data_dir / f"{fname}.csv", encoding="utf-8-sig")
```
