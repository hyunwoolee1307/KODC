```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 0. 기본 설정 및 데이터 준비
# 마스터 표준 수심 리스트 (WOA 기준)
standard_depths = [0, 10, 20, 30, 50, 75, 100, 125, 150, 200, 250, 300, 400, 500]

# 정점별 실제 최대 관측 수심 매핑 (1번 정점 포함 1~4 전체)
max_depths = {1: 70, 2: 130, 3: 150, 4: 143}
stations = [1, 2, 3, 4]

# 2개월 주기 정기 날짜 인덱스 생성 (1962-02-15 ~ 2025-12-15)
new_time_index = pd.date_range(start="1962-02-15", end="2025-12-15", freq="2MS") + pd.DateOffset(days=14)

# 모든 정점의 최종 Anomaly 결과를 저장할 마스터 딕셔너리
all_anomaly_results = {}

# 한글 깨짐 방지 설정
plt.rcParams['axes.unicode_minus'] = False

# 1. Station 1, 2, 3, 4 통합 루프 실행
for sta in stations:
    print(f"=========================================")
    print(f"▶ Station {sta} 처리 및 시각화 시작")
    print(f"=========================================")
    
    # 해당 정점 데이터 필터링 및 시간 타입 변환
    df_sta = line_208.loc[line_208["sta_cde"] == sta, :].copy()
    df_sta["obs_dtm"] = pd.to_datetime(df_sta["obs_dtm"])
    time_index = df_sta["obs_dtm"].unique()
    
    # [동적 설정] 해당 정점의 최대 수심에 맞는 표준 수심 리스트 생성
    interp_depth = [d for d in standard_depths if d <= max_depths[sta]]
    
    # 2. 수심축(Vertical) 선형 보간 수행
    ti = []
    for t in time_index:
        station_data = df_sta.loc[df_sta["obs_dtm"] == t]
        d = station_data["wtr_dep"]
        t_val = station_data["wtr_tmp"]
        
        # 데이터가 부족하거나 빈 경우를 대비해 보간 수행
        interp_t = np.interp(interp_depth, d, t_val)
        ti.append(interp_t)
        
    interp_df = pd.DataFrame(ti, index=time_index, columns=interp_depth)
    
    # 3. 시간축(Temporal) 보간 수행 (불규칙 시계열 -> 정기 2개월 주기)
    total_index = interp_df.index.union(new_time_index)
    df_filled = interp_df.reindex(total_index).sort_index().interpolate(method="time")
    final_interp_df = df_filled.loc[new_time_index]
    
    # 4. Climatology (1991~2020 평년값) 및 Anomaly (기후 편차) 계산
    reference_period = final_interp_df.loc["1991":"2020"]
    climatology = reference_period.groupby(reference_period.index.month).mean()
    
    anomaly_df = final_interp_df.groupby(final_interp_df.index.month, group_keys=False).apply(
        lambda x: x - climatology.loc[x.name]
    ).sort_index()
    
    # 나중에 분석에 쓸 수 있도록 마스터 딕셔너리에 저장
    all_anomaly_results[sta] = anomaly_df
    
    # 5. 수심별 기후 편차(Anomaly) 시각화
    plot_data = anomaly_df.T
    X, Y = plot_data.columns, plot_data.index
    
    plt.figure(figsize=(14, 5))
    
    # 0(평년) 기준 대칭 컬러맵 설정을 위한 절대값 최댓값 계산
    max_val = np.nanmax(np.abs(plot_data.values)) if not np.all(np.isnan(plot_data.values)) else 5.0
    mesh = plt.contourf(X, Y, plot_data.values, cmap="bwr", levels=20, vmin=-max_val, vmax=max_val)
    
    # 차트 레이아웃 및 디자인
    plt.title(f"Line 208 - Station {sta}: Water Temperature Anomalies (1962-2025)\nBase Period: 1991-2020", fontsize=14, fontweight='bold', pad=10)
    plt.xlabel("Observation Date", fontsize=11)
    plt.ylabel("Depth (m)", fontsize=11)
    
    # 수심 반전 (표층이 위로, 심층이 아래로)
    plt.gca().invert_yaxis()
    
    # 컬러바 설정
    cbar = plt.colorbar(mesh)
    cbar.set_label("Temperature Anomaly (°C)", fontsize=10)
    
    plt.tight_layout()
    plt.show()

print("✔ 모든 정점의 처리가 완료되었습니다!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from eofs.standard import Eof

# 1. 4개 정점의 데이터를 하나의 거대한 데이터프레임으로 결합 (공간축 통합)
combined_dfs = []
for sta, df in all_anomaly_results.items():
    renamed_df = df.rename(columns=lambda x: f"Sta{sta}_{x}m")
    combined_dfs.append(renamed_df)

# 시간 인덱스를 기준으로 가로로 병합
eof_input_df = pd.concat(combined_dfs, axis=1)

# 2. eofs 패키지를 이용한 EOF 분석 수행
# eofs 패키지는 (time, space) 형태의 2차원 numpy 배열을 입력으로 받습니다.
data_matrix = eof_input_df.values

# solver 객체 생성 및 분석 실행
solver = Eof(data_matrix)

# 4번 모드까지 결과 추출
# comps: 공간 패턴 (EOF maps), pcs: 시간 변동성 (PC time-series)
# eofs() 결과의 구조는 (mode, space)입니다.
comps = solver.eofs(neofs=4)
pcs = solver.pcs(npcs=4)
var_fractions = solver.varianceFraction()[:4] # 기여율

# 3. 시각화: 4개 모드의 PC Time-series 및 EOF 공간 가중치 출력
fig, axes = plt.subplots(4, 2, figsize=(15, 12), sharex='col')

for i in range(4):
    mode_idx = i  # 0부터 시작하는 인덱스
    mode_num = i + 1
    exp_var = var_fractions[mode_idx] * 100
    
    # --- 좌측: PC 시계열 그래프 ---
    ax_pc = axes[i, 0]
    ax_pc.plot(eof_input_df.index, pcs[:, mode_idx], color='black', lw=1)
    ax_pc.axhline(0, color='gray', linestyle='--', lw=0.8)
    ax_pc.set_title(f"PC Time-series (Mode {mode_num} - {exp_var:.1f}%)", fontsize=11, fontweight='bold')
    ax_pc.grid(True, alpha=0.3)
    
    # --- 우측: EOF Map (정점별 수심에 따른 라인 플롯) ---
    ax_eof = axes[i, 1]
    
    # 해당 모드의 공간 패턴 배열을 Pandas Series로 임시 래핑하여 매핑을 쉽게 만듭니다.
    mode_comps_series = pd.Series(comps[mode_idx], index=eof_input_df.columns)
    
    # 정점별로 선을 따로 그려 시각적으로 분리
    for sta in [1, 2, 3, 4]:
        sta_cols = [col for col in eof_input_df.columns if col.startswith(f"Sta{sta}_")]
        depths = [int(col.split('_')[1].replace('m', '')) for col in sta_cols]
        values = mode_comps_series[sta_cols].values
        
        # x축: EOF 가중치, y축: 수심 (Vertical Profile 형태)
        ax_eof.plot(values, depths, marker='o', label=f"Station {sta}", lw=1.5)
        
    ax_eof.axvline(0, color='gray', linestyle='--', lw=0.8)
    ax_eof.set_title(f"EOF Spatial Pattern (Mode {mode_num})", fontsize=11, fontweight='bold')
    ax_eof.set_ylabel("Depth (m)")
    ax_eof.invert_yaxis()  # 수심 축 반전 (표층이 위로)
    ax_eof.grid(True, alpha=0.3)
    
    if i == 0:
        ax_eof.legend(loc="lower right", fontsize=9)

# 최하단 그래프에만 축 레이블 추가
axes[3, 0].set_xlabel("Observation Date")
axes[3, 1].set_xlabel("EOF Amplitude")

plt.tight_layout()
plt.show()

print("✔ eofs 패키지 기반 EOF 분석 및 시각화 완료")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# --- 1. Station별 경위도 최빈값(Mode) 추출 및 거리(km) 계산 ---
# 하버사인 공식 (위경도 좌표로 두 점 사이의 대권거리 구하기)
def haversine_distance(lat1, lon1, lat2, lon2):
    r = 6371.0 # 지구 반지름 (km)
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    a = np.sin(delta_phi/2)**2 + np.cos(phi1)*np.cos(phi2)*np.sin(delta_lambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return r * c

# 정점별 최빈값 위경도 추출
sta_coords = {}
for sta in [1, 2, 3, 4]:
    df_sta = line_208[line_208["sta_cde"] == sta]
    # 최빈값(mode)의 첫 번째 값 선택
    lat_mode = df_sta["lat"].mode()[0]
    lon_mode = df_sta["lon"].mode()[0]
    sta_coords[sta] = (lat_mode, lon_mode)

# Station 1을 기준(0km)으로 한 각 정점의 누적 거리 계산
sta_distances = {1: 0.0}
lat1, lon1 = sta_coords[1]
for sta in [2, 3, 4]:
    lat2, lon2 = sta_coords[sta]
    sta_distances[sta] = haversine_distance(lat1, lon1, lat2, lon2)

print("--- 정점별 기준(Sta 1) 거리 (km) ---")
for sta, dist in sta_distances.items():
    print(f"Station {sta}: {dist:.2f} km (Lat: {sta_coords[sta][0]}, Lon: {sta_coords[sta][1]})")


# --- 2. 2차원 공간 단면 격자 구성을 위한 데이터 재배열 ---
# eofs 분석 결과인 comps(shape: mode x space)를 다시 데이터프레임 컬럼 구조와 매핑
# eof_input_df는 이전 셀에서 결합한 데이터프레임입니다.
all_features = []
for col in eof_input_df.columns:
    # 예: 'Sta1_10m' -> sta=1, depth=10
    parts = col.split('_')
    sta = int(parts[0].replace('Sta', ''))
    depth = int(parts[1].replace('m', ''))
    all_features.append({'column': col, 'sta': sta, 'depth': depth, 'dist': sta_distances[sta]})

df_features = pd.DataFrame(all_features)


# --- 3. 2차원 단면 시각화 스크립트 (PC 시계열 + 2D EOF Map) ---
fig, axes = plt.subplots(4, 2, figsize=(16, 14), gridspec_kw={'width_ratios': [1.2, 1]})
plt.rcParams['axes.unicode_minus'] = False

# 시각화용 공통 격자 생성 (X: 거리 0~최대, Y: 수심 0~150m)
max_dist = max(sta_distances.values())
grid_x, grid_y = np.mgrid[0:max_dist:100j, 0:150:100j]

for i in range(4):
    mode_idx = i
    mode_num = i + 1
    exp_var = var_fractions[mode_idx] * 100
    
    # 1) 좌측: PC Time-series
    ax_pc = axes[i, 0]
    ax_pc.plot(eof_input_df.index, pcs[:, mode_idx], color='black', lw=1)
    ax_pc.axhline(0, color='gray', linestyle='--', lw=0.8)
    ax_pc.set_title(f"PC Time-series (Mode {mode_num} - {exp_var:.1f}%)", fontsize=11, fontweight='bold')
    ax_pc.grid(True, alpha=0.3)
    if i == 3:
        ax_pc.set_xlabel("Observation Date")
        
    # 2) 우측: 2D EOF Spatial Cross-section (X: Distance, Y: Depth)
    ax_eof = axes[i, 1]
    
    # 현재 모드의 EOF 가중치 값 매칭
    df_features['eof_val'] = comps[mode_idx]
    
    # 불규칙한 (Distance, Depth) 포인트를 2차원 바둑판 격자로 선형 보간
    points = df_features[['dist', 'depth']].values
    values = df_features['eof_val'].values
    grid_z = griddata(points, values, (grid_x, grid_y), method='linear')
    
    # 등고선 채우기 (대칭 컬러바를 위해 절대값 최댓값 기준 vmin, vmax 설정)
    max_val = np.nanmax(np.abs(values))
    mesh = ax_eof.contourf(grid_x, grid_y, grid_z, cmap="bwr", levels=30, vmin=-max_val, vmax=max_val)
    
    # 실제 관측 포인트 위치 표시 (데이터 분포 확인용 점)
    ax_eof.scatter(df_features['dist'], df_features['depth'], c='black', s=10, alpha=0.6, label='Obs Points' if i==0 else "")
    
    # 상단에 Station 위치를 알 수 있도록 X축 상단 레이블 대용 점선 추가
    for sta, dist in sta_distances.items():
        ax_eof.axvline(dist, color='gray', linestyle=':', alpha=0.5)
        if i == 0:
            ax_eof.text(dist, -5, f"Sta {sta}", ha='center', fontsize=9, fontweight='bold')

    ax_eof.set_title(f"EOF Spatial Pattern Section (Mode {mode_num})", fontsize=11, fontweight='bold')
    ax_eof.set_ylabel("Depth (m)")
    ax_eof.invert_yaxis() # 수심 축 반전 (표층이 위로)
    
    # 컬러바 설정
    cbar = fig.colorbar(mesh, ax=ax_eof, pad=0.02)
    cbar.set_label("EOF Amplitude", fontsize=9)
    
    if i == 3:
        ax_eof.set_xlabel("Distance from Station 1 (km)")
    if i == 0:
        ax_eof.legend(loc="lower left", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
def process_anomaly_with_filter(dataframe, min_observations=20):
    """
    KODC 데이터셋의 결측 및 날짜 파싱 오류를 방지하면서 
    지정된 개수 이상의 유효 관측치를 가진 정점만 안전하게 추출합니다.
    """
    df_mapped = dataframe.copy()
    
    # 1. 컬럼명 매핑
    df_mapped = df_mapped.rename(columns={
        'Line': 'sln_cde', 'Station': 'sta_cde',
        'Latitude': 'lat', 'Longitude': 'lon',
        'Date(KST)': 'obs_dtm', 'Depth(m)': 'wtr_dep',
        'Water Temp(°C)': 'wtr_tmp'
    })
    
    # 2. 날짜 파싱 (errors='coerce'를 주어 잘못된 포맷은 NaT 처리 후 드롭)
    df_mapped['obs_dtm'] = pd.to_datetime(df_mapped['obs_dtm'], errors='coerce')
    df_mapped = df_mapped.dropna(subset=['obs_dtm'])
    
    # 3. 1982 ~ 2025년 기간 필터링 (안전한 String/Timestamp 비교)
    df_mapped = df_mapped[(df_mapped['obs_dtm'] >= pd.Timestamp("1982-01-01")) & 
                          (df_mapped['obs_dtm'] <= pd.Timestamp("2025-12-31"))]
    
    # 수온 유효 데이터가 없는 행도 사전 제거
    df_mapped = df_mapped.dropna(subset=['wtr_tmp'])
    
    print(f"▶ [로그] 기간 필터링 후 남은 전체 데이터 로우 수: {len(df_mapped)}")
    
    all_stations = []
    meta_list = []
    
    # 4. 정점별 루프
    for (sln, sta), group in df_mapped.groupby(['sln_cde', 'sta_cde']):
        # [수정] 단순 고유 관측일(날짜 수) 개수로 직관적이고 안전하게 변경
        n_valid_dates = group['obs_dtm'].nunique()
        
        if n_valid_dates < min_observations:
            # 관측치가 부족해서 버려지는 정점 추적용 로그 (필요시 주석 해제)
            # print(f"정점 제외: Line {sln} Sta {sta} (유효 관측일 수: {n_valid_dates}개)")
            continue
            
        lat_m = group['lat'].mode()[0]
        lon_m = group['lon'].mode()[0]
        t_idx = group['obs_dtm'].unique()
        
        max_d = group['wtr_dep'].max()
        interp_d = standard_depths[standard_depths <= max_d]
        if len(interp_d) == 0: 
            continue
            
        # 수심 보간 세션
        ti = []
        for t in t_idx:
            sub = group[group['obs_dtm'] == t]
            if sub.empty or sub['wtr_tmp'].isna().all():
                ti.append(np.full(len(interp_d), np.nan))
                continue
            sub_s = sub.dropna(subset=['wtr_tmp']).sort_values('wtr_dep')
            if sub_s.empty:
                ti.append(np.full(len(interp_d), np.nan))
                continue
            ti.append(np.interp(interp_d, sub_s['wtr_dep'], sub_s['wtr_tmp'], left=np.nan, right=np.nan))
            
        interp_df = pd.DataFrame(ti, index=t_idx, columns=interp_d)
        
        # 시간 보간 세션
        df_f = interp_df.reindex(interp_df.index.union(new_time_index)).sort_index().interpolate(method="time")
        final_df = df_f.loc[new_time_index]
        
        # 1991~2020 평년값 기준 Anomaly 계산
        ref = final_df.loc["1991":"2020"]
        if ref.empty: 
            continue
        clim = ref.groupby(ref.index.month).mean()
        anom = final_df.groupby(final_df.index.month, group_keys=False).apply(
            lambda x: x - clim.loc[x.name] if x.name in clim.index else np.nan
        ).sort_index()
        
        for d in interp_d:
            col = f"Sln{int(sln)}_Sta{int(sta)}_{d}m"
            all_stations.append(anom[d].rename(col))
            meta_list.append({'column': col, 'sln': sln, 'sta': sta, 'lat': lat_m, 'lon': lon_m, 'depth': d})
            
    if not all_stations:
        raise ValueError(f"[에러] min_observations={min_observations} 조건을 만족하는 정점이 최종 연산 단계에서 전멸했습니다.\n"
                         f"데이터프레임의 'Line' 혹은 'Date(KST)' 형식을 다시 확인하세요.")
        
    input_df = pd.concat(all_stations, axis=1).dropna(how='all').fillna(0)
    return input_df, pd.DataFrame(meta_list)
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from eofs.standard import Eof

# =========================================================================
# 1. 환경 설정 및 데이터 전처리 함수 정의 (1982 ~ 2025 타겟팅)
# =========================================================================
plt.rcParams['axes.unicode_minus'] = False

# 모든 마스터 레벨 수심 정의 (14개 레이어 전체 대상 시각화)
standard_depths = np.array([0, 10, 20, 30, 50, 75, 100, 125, 150, 200, 250, 300, 400, 500])

# 출력 디렉토리 생성
output_dir = "./eof_composite_results"
os.makedirs(output_dir, exist_ok=True)

# 1982년부터 2025년까지 2개월 주기 정규 타임 인덱스 생성
new_time_index = pd.date_range(start="1982-02-15", end="2025-12-15", freq="2MS") + pd.DateOffset(days=14)

def process_anomaly_with_filter(dataframe, min_observations=132):
    """
    KODC 데이터셋의 결측 및 날짜 파싱 오류를 방지하면서 
    Nyquist 기준(최대 264개 중 절반인 132개) 이상의 유효 관측치를 가진 정점만 추출합니다.
    """
    df_mapped = dataframe.copy()
    
    # 컬럼명 매핑
    df_mapped = df_mapped.rename(columns={
        'Line': 'sln_cde', 'Station': 'sta_cde',
        'Latitude': 'lat', 'Longitude': 'lon',
        'Date(KST)': 'obs_dtm', 'Depth(m)': 'wtr_dep',
        'Water Temp(°C)': 'wtr_tmp'
    })
    
    # 날짜 안전하게 파싱 및 결측치 제거
    df_mapped['obs_dtm'] = pd.to_datetime(df_mapped['obs_dtm'], errors='coerce')
    df_mapped = df_mapped.dropna(subset=['obs_dtm', 'wtr_tmp'])
    
    # [FIX] 단순 문자열 비교 대신 pd.Timestamp 객체 비교로 타임 플래그 에러 완전 방지
    start_date = pd.Timestamp("1982-01-01")
    end_date = pd.Timestamp("2025-12-31")
    df_mapped = df_mapped[(df_mapped['obs_dtm'] >= start_date) & (df_mapped['obs_dtm'] <= end_date)]
    
    print(f"▶ [로그] 1982~2025 기간 필터링 후 남은 데이터 로우 수: {len(df_mapped)}개")
    
    all_stations = []
    meta_list = []
    
    # 정점별 루프
    for (sln, sta), group in df_mapped.groupby(['sln_cde', 'sta_cde']):
        # 고유 관측일(날짜) 개수 추출
        n_valid_dates = group['obs_dtm'].nunique()
        
        # Nyquist 주파수 기준 미달 정점 스크리닝
        if n_valid_dates < min_observations:
            continue
            
        # 생존한 정점 로그 출력
        print(f"   └ [통과] Line {int(sln)} Sta {int(sta)}: 유효 관측일 {n_valid_dates}개 확보 완료")
            
        lat_m = group['lat'].mode()[0]
        lon_m = group['lon'].mode()[0]
        t_idx = group['obs_dtm'].unique()
        
        max_d = group['wtr_dep'].max()
        interp_d = standard_depths[standard_depths <= max_d]
        if len(interp_d) == 0: 
            continue
            
        # 수심 보간
        ti = []
        for t in t_idx:
            sub = group[group['obs_dtm'] == t]
            if sub.empty or sub['wtr_tmp'].isna().all():
                ti.append(np.full(len(interp_d), np.nan))
                continue
            sub_s = sub.dropna(subset=['wtr_tmp']).sort_values('wtr_dep')
            if sub_s.empty:
                ti.append(np.full(len(interp_d), np.nan))
                continue
            ti.append(np.interp(interp_d, sub_s['wtr_dep'], sub_s['wtr_tmp'], left=np.nan, right=np.nan))
            
        interp_df = pd.DataFrame(ti, index=t_idx, columns=interp_d)
        
        # 시간 보간
        df_f = interp_df.reindex(interp_df.index.union(new_time_index)).sort_index().interpolate(method="time")
        final_df = df_f.loc[new_time_index]
        
        # 1991~2020 평년값 기준 Anomaly 계산
        ref = final_df.loc["1991":"2020"]
        if ref.empty: 
            continue
        clim = ref.groupby(ref.index.month).mean()
        anom = final_df.groupby(final_df.index.month, group_keys=False).apply(
            lambda x: x - clim.loc[x.name] if x.name in clim.index else np.nan
        ).sort_index()
        
        for d in interp_d:
            col = f"Sln{int(sln)}_Sta{int(sta)}_{d}m"
            all_stations.append(anom[d].rename(col))
            meta_list.append({'column': col, 'sln': sln, 'sta': sta, 'lat': lat_m, 'lon': lon_m, 'depth': d})
            
    if not all_stations:
        raise ValueError(f"[에러] min_observations={min_observations} 조건을 만족하는 정점이 없습니다.")
        
    input_df = pd.concat(all_stations, axis=1).dropna(how='all').fillna(0)
    return input_df, pd.DataFrame(meta_list)

# =========================================================================
# 2. 전처리 파이프라인 실행 및 Line 208 기준 EOF 분석 (기준: 240일 상향)
# =========================================================================
# [FIX] 유효 관측일 기준을 240일로 상향 조정 (Discontinued 정점 완벽 배제)
target_min_obs = 240

print(f"1. Line 208 데이터 필터링 및 Anomaly 처리 중 (최소 {target_min_obs}일)...")
line_208_data = dfs[dfs['sln_cde'] == 208] if 'sln_cde' in dfs.columns else dfs[dfs['Line'] == 208]
l208_anom_df, l208_meta = process_anomaly_with_filter(line_208_data, min_observations=target_min_obs)

print(f"\n2. 전체 해양 데이터 필터링 및 Anomaly 처리 중 (최소 {target_min_obs}일)...")
all_anom_df, all_meta = process_anomaly_with_filter(dfs, min_observations=target_min_obs)

# 공통 시간축 연동
common_times = l208_anom_df.index.intersection(all_anom_df.index)
l208_anom_df = l208_anom_df.loc[common_times]
all_anom_df = all_anom_df.loc[common_times]

print("\n3. Line 208 EOF(Mode 1~4) 분석 수행 중...")
solver_208 = Eof(l208_anom_df.values)
pcs_208 = solver_208.pcs(npcs=4) 
var_frac_208 = solver_208.varianceFraction()[:4]

# =========================================================================
# 3. 모든 독립 그림 파일별 대륙 마스킹 및 개별 시각화 (Aspect & Suptitle 픽스)
# =========================================================================
print(f"\n4. 독립형 시각화 렌더링 및 이미지 저장 시작 (저장 경로: {output_dir})...")
proj = ccrs.PlateCarree()

lon_min, lon_max = all_meta['lon'].min() - 0.5, all_meta['lon'].max() + 0.5
lat_min, lat_max = all_meta['lat'].min() - 0.5, all_meta['lat'].max() + 0.5
grid_lon, grid_lat = np.mgrid[lon_min:lon_max:200j, lat_min:lat_max:200j]

# Mode 1~4 루프
for mode_idx in range(4):
    mode_num = mode_idx + 1
    exp_var = var_frac_208[mode_idx] * 100
    
    pc_mode = pd.Series(pcs_208[:, mode_idx], index=common_times)
    pc_mode_std = (pc_mode - pc_mode.mean()) / pc_mode.std()
    
    pos_dates = pc_mode_std[pc_mode_std >= 1.0].index
    if len(pos_dates) > 0:
        composite_pos = all_anom_df.loc[all_anom_df.index.isin(pos_dates)].mean()
    else:
        composite_pos = all_anom_df.mean()
        
    all_meta['composite_val'] = composite_pos.reindex(all_meta['column']).values
    
    # 수심별 루프
    for d in standard_depths:
        df_d = all_meta[(all_meta['depth'] == d) & (all_meta['composite_val'].notna())]
        if df_d.empty or len(df_d) < 3:
            continue
            
        fig, ax = plt.subplots(figsize=(8, 9.5), subplot_kw={'projection': proj}, layout="constrained")
        ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=proj)
        
        grid_z = griddata(
            points=df_d[['lon', 'lat']].values, 
            values=df_d['composite_val'].values, 
            xi=(grid_lon, grid_lat), 
            method='linear'
        )
            
        mesh = ax.contourf(grid_lon, grid_lat, grid_z, cmap="bwr", levels=contour_levels, 
                           vmin=cbar_min, vmax=cbar_max, transform=proj, extend="both")
        
        ax.scatter(df_d['lon'], df_d['lat'], c='black', s=6, alpha=0.5, transform=proj, zorder=3)
        ax.add_feature(cfeature.LAND, facecolor='#dddddd', edgecolor='none', zorder=2)
        ax.add_feature(cfeature.COASTLINE, linewidth=0.8, edgecolor='#444444', zorder=2)
        
        # [FIX] ax.set_title을 제거하고, fig.suptitle로 전환하여 폰트 24pt/bold 적용
        fig.suptitle(fr"Mode {mode_num} Composite Anomaly [{int(d)}m]", 
                     fontsize=24, fontweight='bold', y=0.98)
        
        gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.3, linestyle='--')
        gl.top_labels = False
        gl.right_labels = False
        
        # [FIX] aspect=50 설정으로 컬러바 두께 최적화
        cbar = fig.colorbar(mesh, ax=ax, orientation='vertical', pad=0.03, shrink=0.7, aspect=50)
        cbar.set_label("Temperature Anomaly Composite (°C)", fontsize=12, labelpad=10)
        
        major_ticks = np.arange(cbar_min, cbar_max + 0.01, 0.5)
        cbar.set_ticks(major_ticks)
        cbar.ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.1f'))
        
        minor_ticks = np.arange(cbar_min, cbar_max + 0.01, 0.1)
        cbar.ax.yaxis.set_ticks(minor_ticks, minor=True)
        
        file_name = f"composite_mode{mode_num}_depth_{int(d)}m.png"
        plt.savefig(os.path.join(output_dir, file_name), dpi=200)
        plt.close(fig)

print(f"\n✔ [완료] fig.suptitle 적용 및 컬러바 aspect=50 설정으로 시각화가 완료되었습니다!")
```
